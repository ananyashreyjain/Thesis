```python
import pandas as pd
import numpy as  np
import tensorflow as tf
```

```python
data=pd.read_csv("Design1a_Para2_S21.csv")
```

```python
data.head()
```

```python
remove_columns=[]
for index in range(0, data.shape[1]):
    if (data.iloc[:, index]==-1*np.inf).all():
        remove_columns.append(data.columns[index])
data=data.drop(remove_columns, axis=1)
```

```python
def input_parameters():
    output=np.empty((0,4), float)
    for columns in data.columns[1:]:
        params=columns.split(' ')[3:]
        param=[]
        for vals in params:
            param.append(float(vals.split('=')[1][1:-3]))
        param=np.array([param])
        output=np.concatenate((output, param), axis=0)
    return output
```
